```python
# Linear Interpolation
#T1South	Developed land/total
import numpy
import matplotlib.pyplot as plt
from scipy import interpolate

x = [2001,2004,2006,2008,2011,2013,2016,2019]
y = [0.074051,0.07578,0.081054,0.082366,0.083079,0.083587,0.084386,0.08557]
f = interpolate.interp1d(x, y)
print(f(range(2001,2020)))
xnew = range(2001,2020)
ynew = f(xnew)   # use interpolation function returned by `interp1d`
plt.plot(x, y, 'o', xnew, ynew, '-')
plt.show()
```

```python
# Polynomial Extrapolation
# T1South	Developed land/total

x = [2001,2004,2006,2008,2011,2013,2016,2019]
y = [0.074051,0.07578,0.081054,0.082366,0.083079,0.083587,0.084386,0.08557]

mymodel = numpy.poly1d(numpy.polyfit(x, y, 3))

myline = numpy.linspace(2001, 2019)

plt.scatter(x, y)
plt.plot(myline, mymodel(myline))
plt.show()

mymodel(range(1997,2021))
```
